```python
import pandas as pd
```

```python
df1 = pd.read_csv("Books Details.csv")
df2 = pd.read_csv("Individual Book details.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.DataFrame()
```

```python
df['Title']               = df1["Title"]
df["UPC"]                 = df2["UPC"]
df["Category"]            = df2["Type"]
df["Price_excluding_tax"] = df2["Price_excluding_tax"]
df["Price_including_tax"] = df2["Price_including_tax"]
df["Tax"]                 = df2["Tax"]
df["Rating"]              = df1["Rating"]
df["Reviews"]             = df2["Reviews"]
df["Quantity"]            = df2["Quantity"]
df["Stocks"]               = df1["Stocks"]

df["Book_link"]           = df1["Book_link"]
df["Image_link"]          = df1["Image_link"]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv("All book details.csv", index = False)
```
